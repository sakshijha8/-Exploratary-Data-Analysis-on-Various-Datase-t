This article is a list of the top 50 accounts with most followers on the social platform Instagram. As of May 2022, the most-followed individual is footballer Cristiano Ronaldo with over 441 million followers, and the most-followed woman is reality television star Kylie Jenner with over 336 million followers. Instagram's own brand account on the platform is the most-followed account overall, with over 504 million followers. National Geographic is the second most-followed brand account, with over 221 million followers. A total of 34 accounts have exceeded 100 million followers on the platform, with 16 of those having exceeded 200 million followers, 8 of those having exceeded 300 million, 2 exceeding 400 million, and 1 exceeding 500 million.

# Importing Libraries And Loading the Dataset:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)
```

```python
data = pd.read_csv('Top most followed instagram accounts.csv',encoding='unicode_escape')
```

### Let's build a graph of the distribution of a categorical feature.

```python
plt.figure(figsize=(20,5))
sns.set_style('whitegrid')
sns.countplot(data['Country/Continent'])
plt.xticks(rotation=90)
```

The largest number of top users are located in the USA. The reason for the greatest popularity of users from the United States lies in the active socio-cultural position of this country, most of the stars of modern cinema live in the United States. India is in second place. The reason lies in the large population. The absence of China in this ranking is due to the high level of state censorship and the high prevalence of local social networks.

```python
plt.figure(figsize=(25,6))
sns.set_style('whitegrid')
sns.countplot(data['Profession/Activity'])
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(50,8))
sns.set_style('whitegrid')
sns.countplot(data['Username'])
plt.xticks(rotation=90)
```

### Let's evaluate continuous feature.

```python
sns.histplot(data["Followers (millions)"], stat='density',color='green')
sns.kdeplot(data["Followers (millions)"], color='black')
```

### Apply log transformation to the attribute:

```python
data['Followers (millions)Log'] = np.log(data['Followers (millions)']+1)
sns.histplot(data["Followers (millions)Log"], stat='density')
sns.kdeplot(data["Followers (millions)Log"], color='black')
```
