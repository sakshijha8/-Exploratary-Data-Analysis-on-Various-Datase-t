This dataset contains information about all the listed restaurants in Hyderabad. This dataset can be utilized to build recommendation system, analyze data trend, predict food price and much more. 

**Dataset column details are as follows:**

* links: Contains links to the order page of the restaurants.
* names: Name of the restaurants.
* ratings: Average of the rating given by the all the customers.
* cuisine: Cuisine served by the restaurants.
* price for one: Cost of the food for one person in INR.

# Exploratary Data Analysis:

Let's build a graph of the distribution of a categorical feature.¶

```python
plt.figure(figsize=(100,8))
sns.set_style('whitegrid')
sns.countplot(data['names'])
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(100,8))
sns.set_style('whitegrid')
sns.countplot(data['ratings'])
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(60,5))
sns.set_style('whitegrid')
sns.countplot(data['cuisine'])
plt.xticks(rotation=90)
```

## Let's evaluate continuous feature.

```python
sns.histplot(data["price for one"], stat='density',color='green')
sns.kdeplot(data["price for one"], color='black')
```

```python
#look at how target is distributed among variables
sns.pairplot(data)
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(data , hue = 'ratings')
```

## Apply log transformation to the attribute price for one:

```python
data['price for oneLog'] = np.log(data['price for one']+1)
sns.histplot(data["price for oneLog"], stat='density')
sns.kdeplot(data["price for oneLog"], color='black')
```

```python
sns.boxplot(data['price for one'],color = 'green')
```
